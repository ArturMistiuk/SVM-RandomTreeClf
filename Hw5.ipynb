```python
import os
import pandas as pd
```

```python
combined_data = pd.DataFrame()
accelerometer_data_folder = 'data'
```

```python
for activity_folder in os.listdir(accelerometer_data_folder):
    activity_path = os.path.join(accelerometer_data_folder, activity_folder)
    for file in os.listdir(activity_path):
        if file.endswith('.csv'):
            file_path = os.path.join(activity_path, file)
            df = pd.read_csv(file_path)
            df['Activity'] = activity_folder
            combined_data = pd.concat([combined_data, df])
```

```python
combined_data.head()
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier, LazyRegressor
from sklearn.model_selection import train_test_split
```

```python
X = combined_data[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']]
y = combined_data['Activity']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
clf = LazyClassifier()
```

```python
models = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models[0]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.tree import ExtraTreeClassifier
```

```python
svm_clf = svm.SVC()
svm_clf.fit(X_train, y_train)
svm_y_pred = svm_clf.predict(X_test)
```

```python
random_f_clf = RandomForestClassifier()
random_f_clf.fit(X_train, y_train)
rfc_y_pred = random_f_clf.predict(X_test)
```

```python
extra_t_clf = ExtraTreeClassifier()
extra_t_clf.fit(X_train, y_train)
etc_y_pred = extra_t_clf.predict(X_test)
```

```python
svm_report = classification_report(y_test, svm_y_pred)
print('SVM Report:')
print(svm_report)
rfc_report = classification_report(y_test, rfc_y_pred)
print('Random Forest Classifier Report:')
print(rfc_report)
etc_report = classification_report(y_test, etc_y_pred)
print('Extra Tree Classifier Report:')
print(etc_report)
```

### Висновки:
Алгоритм Random Forest виявляє кращі результати точностi. Extra Tree є швидшим i таким самим по точностi. SVM показує найгiршi результати.
